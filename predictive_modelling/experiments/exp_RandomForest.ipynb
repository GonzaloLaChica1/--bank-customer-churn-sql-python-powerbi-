<h2><strong>Logistic Regression</strong></h2>

**Load data and Libraries**




```python
import warnings
warnings.filterwarnings("ignore")
import joblib
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from evaluation_script import evaluate_model

```

```python
#LOAD PROCESSED DATA 

data= joblib.load('../processed_data/dataset_bundle.pkl')
X_train= data['X_train']
y_train=data['Y_train']
X_test=data['X_test']
y_test=data['Y_test']
```

**Find the best Parameters**


```python
#declare parameters

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 10, None],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt"],
}
```

```python
from sklearn.ensemble import RandomForestClassifier

#generate search space
RanFor_Grid = GridSearchCV(
    RandomForestClassifier(class_weight="balanced", random_state=200),
    param_grid,
    cv=5,
    refit=True,
    scoring="f1",
    n_jobs=-1,
    verbose=1
)
```

FIT BEST MODEL ON TRAINING 

```python
RanFor_Grid.fit(X_train, y_train)
best_model = RanFor_Grid.best_estimator_

results = evaluate_model(best_model, X_test, y_test)
results
```
